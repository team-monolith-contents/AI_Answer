**[문제1] 의사 결정 트리를 활용하기 좋은 문제 상황을 적어봅시다.**

- 질병 예측 : 병원에서 환자에게 이런 조건에 부합되기 때문에 어떤 질병이 의심된다. 
- 대출 가능 여부 판단 : 은행에서 이런 이유들 때문에 대출이 제한되었다 




**[문제2] 문제1에서 생각한 문제 상황에 의사 결정 트리를 이용할 때 사용될 수 있는 노드의 분류 기준을 예상해 봅시다.**

- 질병 예측 : 최근에 5kg 이상의 변화가 있다. 식사 후 소화가 되지 않는다. 
- 대출 가능 여부 판단 : 최근 일년 이내에 신용카드 대금을 연체한 적이 있다. 




**[문제3] 다음 환자들의 특징을 이용해 진단명에 맞게 잘 분류할 수 있는 가장 깊이가 얕은 의사 결정 트리를 만들어 봅시다.**


(예시)
> 목 쉼 없음
>> 열 37도 미만 -------- 감기
>>
>> 열 37도 이상
>>> 기침 2일 미만 -------- 인후염
>>>
>>> 기침 2일 이상 -------- 독감
>>
> 목 쉼 있음
>> 기침 2일 미만 -------- 갑상샘 저하증
>>
>> 기침 2일 이상
>>> 열 37도 미만 -------- 성대 감염
>>>
>>> 열 37도 이상 -------- 폐렴





**[문제4] 학습된 의사 결정 트리의 분류 기준을 보고 유추할 수 있는 각 질병의 특징을 적어봅시다.**

- 성대 감염 : 열이 나지 않지만 목이 쉬고 기침이 있음
- 갑상샘 저하증 : 기침이 없지만 열이 조금 있고 목 쉰 기간이 김
- 폐렴 : 열이 매우 높고 목 쉼, 기침 증상이 모두 있음
- 인후염 : 목이 쉬거나 기침은 없지만 열이 있음
- 감기 : 열과 목 쉼 증상이 없고 기침이 있음
- 독감 : 목 쉼 증상이 없고 열이 매우 높으며 기침이 있음





**[문제5] 학습된 의사 결정 트리에 새로운 환자데이터를 넣어보고 예측한 진단명을 적어봅시다.**

(예시) {'체온': 38.5, '목 통증 여부': 0, '목 쉼 기간': 0, '기침 기간': 1} 의 데이터를 가진 환자는 '폐렴'으로 분류되었습니다.




**[문제6] 학습된 의사 결정 트리의 분류 기준과 결과에서 찾은 문제점과 데이터 수집 단계로 돌아가 해결할 방법을 적어봅시다.**

(예시) 별 다른 증상이 없을 때에도 '폐렴'으로 분류되었습니다. 정상 환자의 데이터를 추가하면 좋을 것 같습니다.
